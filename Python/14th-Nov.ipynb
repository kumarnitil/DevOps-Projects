List vs numpy
- In list you can perform the row level segregation using the index but you cannot perform the column level 
segregation.
- In Python everything is coming from Library - From Library we get the modules and from the modules we get the functions.
- Library---modules---functions
- pip list will give you the libraries which are already present.
- To perform the column wise operation we can use numpy array.
- numpy will not come with the default libraries and we will have to install it.
- The package will be installed by pip install numpy and then import it.
- import numpy as np



```python
import numpy as np

list1 = [["Name", "Age", "City"],
         ["Alice", 30, "New York"],
         ["Bob", 25, "Los Angeles"],
         ["Charlie", 35, "Chicago"]]

print (list1[1]) # We can extract the second list using index 1,
                 # but we cannot access columns directly using index. 
```

- In the numpy library we have the function "array"


```python
list2 = np.array(list1) # We can convert the list of lists into a NumPy array, which allows us to access columns directly.
print (list2[:, 1]) # We can extract the second column using index 1.

print (list2[1:4, 1]) # We can also extract the second column for all rows except the header.
                     # Before the comma, we specify the range of rows (1 to 4, which includes rows 1, 2, and 3), and after the comma, we specify the column index (1 for the second column).
```
