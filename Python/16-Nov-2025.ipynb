- We use smtplib library to send emails and it is a standard library which means it comes pre-installed.
- Smtp library will be used to login into the gmail and the email library will be used to construct the email body.

```python
import smtplib
from email.mime.text import MIMEText
sender = "aiopsboy@gmail.com"
# You cannot give the password directly, you need to generate an app password for your Gmail account and use it here.
# In case of app password, it does not require captcha verification, so it can be used in scripts and applications.
# Every time you change your Gmail password, you will need to generate a new app password and update it in your script. This is a security measure to protect your account.
password = "jtancrmvykruazyj"
recipient = "kumarnitildevops@gmail.com"
# recipient = ["kumarnitildevops@gmail.com", "anotherrecipient@gmail.com"]  # You can specify multiple recipients by providing a list of email addresses
subject = "Test Email from AiOpsBoy"
body = "This is a test email sent from a Python script using Gmail's SMTP server."
msg = MIMEText(body) # Create a MIMEText object to represent the email message
msg['subject'] = subject
msg['From'] = sender
msg['To'] = recipient
# Use the following format if you have multiple recipients:
# msg['To'] = ", ".join(recipient)  # Join the list of recipients
# Let us send the email using Gmail's SMTP server
# smtp.gmail.com is the SMTP server for Gmail, and 587 is the port number for TLS encryption.
with smtplib.SMTP('smtp.gmail.com', 587) as server:
    server.starttls() # Upgrade to a secure connection using TLS encryption
    server.login(sender, password)  # Log in to the SMTP server using the sender's email address and password
    server.send_message(msg)  # Send the email message
    # send to multiple recipients    # server.send_message(msg, from_addr=sender, to_addrs=recipient)  
    # # If you have multiple recipients, you can specify the sender and recipient addresses explicitly using the from_addr and
    print("Email sent successfully!")
```
